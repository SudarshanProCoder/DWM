```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)

k = 5 

min_val, max_val = data.min(), data.max()
width = (max_val - min_val) / k
bins = [min_val + i * width for i in range(k+1)]

labels = [f"Bin{i+1}" for i in range(k)]
discretized = pd.cut(data, bins=bins, labels=labels, include_lowest=True)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Original Continuous Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
discretized.value_counts().sort_index().plot(kind='bar', color='salmon', edgecolor='black')
plt.title("Discretized Data (Equal Width)")
plt.xlabel("Bins")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
